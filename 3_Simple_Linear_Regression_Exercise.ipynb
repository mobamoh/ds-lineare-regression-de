## Frage 2

Die Datei `utils.csv` in Ihrem data-Ordner enthält Informationen über die durchschnittlichen Nebenkosten für Häuser einer bestimmten Größe und die durchschnittliche monatliche Temperatur.

**(a) Laden Sie die Daten:**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

sns.set_style("ticks")
plt.rcParams["figure.figsize"] = (7, 4)
```

```python
utils = (
    pd.read_csv("data/utils.csv")
    .rename(
        columns={
            "Average Bill": "average_bill",
            "Average Monthly Temperature": "average_monthly_temperature",
        }
    )
)

utils
```

**(b) Erstellen Sie ein Streudiagramm der Daten: durchschnittliche Rechnung vs. durchschnittliche monatliche Temperatur. Welche Variable wäre am sinnvollsten als Antwortvariable?**

```python
sns.scatterplot(
    data=utils,
    x="average_monthly_temperature",
    y="average_bill",
    s=80,
)

plt.xlabel("Average monthly temperature")
plt.ylabel("Average bill");

"""
Die sinnvollste Antwortvariable ist average_bill.
Die Temperatur ist hier der erklaerende Praediktor, und die Rechnung ist das Ergebnis,
das wir vorhersagen moechten.
"""
```

**(c) Schauen Sie sich das obige Diagramm an. Wie würden Sie die Beziehung zwischen den Variablen beschreiben?**

```python
"""
Die Beziehung wirkt nicht besonders gut linear.
Insgesamt gibt es zwar einen leichten negativen Trend, aber das Muster ist eher gekruemmt
und vermutlich saisonal, sodass eine Gerade die Daten nur begrenzt beschreibt.
"""
```

**(d) Trainieren Sie ein einfaches lineares Regressionsmodell mit scikit-learn**

```python
X = utils[["average_monthly_temperature"]]
y = utils["average_bill"]

lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

**(e) Berechnen und interpretieren Sie den Wert von $R^2$.**

```python
y_hat = lin_reg.predict(X)
r_squared = r2_score(y, y_hat)

print("R-squared:", round(r_squared, 3))

"""
Das Modell erklaert nur etwa 12.4% der Varianz der Nebenkosten.
Die lineare Temperatur-Beziehung allein passt also eher schlecht zu diesen Daten.
"""
```

**(f) Geben Sie den Achsenabschnitt und die Steigung/den Koeffizienten aus und interpretieren Sie diese in Begriffen, die ein Hausbesitzer verstehen würde.**

```python
intercept = lin_reg.intercept_
slope = lin_reg.coef_[0]

print("Intercept:", round(intercept, 3))
print("Slope:", round(slope, 3))

"""
Der Achsenabschnitt bedeutet: Bei 0 Grad wuerde das Modell eine durchschnittliche Rechnung
von etwa 143.6 vorhersagen.
Die Steigung bedeutet: Fuer jedes zusaetzliche Grad sinkt die geschaetzte Rechnung im Mittel
um etwa 0.48 Einheiten.
"""
```

```python
prediction_120 = lin_reg.predict(
    pd.DataFrame({"average_monthly_temperature": [120]})
)[0]

print("Predicted average bill at 120 degrees:", round(prediction_120, 3))

"""
Diese Vorhersage ist nicht besonders vernuenftig.
120 Grad liegen weit ausserhalb des beobachteten Bereichs, und die Daten zeigen eher eine
nichtlineare, saisonale Struktur als eine stabile lineare Beziehung.
"""
```
