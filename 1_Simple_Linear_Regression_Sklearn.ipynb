### Was ist lineare Regression?

In der Statistik ist **lineare Regression** ein Ansatz zur Modellierung der Beziehung zwischen einer Antwortvariablen $y$ und einer oder mehreren erklärenden Variablen. Es ist wichtig zu bedenken, dass lineare Regression eine Beziehung approximieren kann und keine Kausalität beweisen kann.

Außerdem macht sie einige **Annahmen**:
1. **Linearität**: Die Zielvariable und die Koeffizienten der erklärenden Variablen sind linear miteinander verbunden.
2. **Fehler mit Mittelwert Null**: Der Mittelwert aller Residuen ist null.
3. **Strikte Exogenität**: Alle erklärenden Variablen sind unkorreliert mit dem Residuum.
4. **Homoskedastizität**: Die Varianz der Residuen ist konstant über alle Niveaus der unabhängigen Variablen.
5. **Keine Multikollinearität**: Alle erklärenden Variablen sind linear unabhängig.

Wir können lineare Regression in zwei Kategorien einteilen:
* **Einfache lineare Regression**: Fälle, in denen wir nur eine erklärende Variable haben
* **Multiple lineare Regression**: Fälle, in denen wir mehr als eine erklärende Variable haben

Die Formel für eine einfache lineare Regression ist gegeben durch

$$
y_i = b_0 + b_1 x_i + e_i
$$  
$$
\text{für i = } 1, \dots, n
$$

wobei $y_i$ der wahre Wert für unsere Antwortvariable ist. Die Formel für die Regressionsgleichung ist gegeben durch

$$
\hat{y}_i = b_0 + b_1 x_i
$$
$$
\text{für i = } 1, \dots, n
$$

wobei $\hat{y}_i$ der vorhergesagte Wert für unsere Antwortvariable ist und $b_0$ und $b_1$ der geschätzte Achsenabschnitt und die Steigung der Geraden sind.

### Wie funktioniert es?

Lineare Regression ist im Grunde nur der schicke Begriff für das Finden der besten Anpassungslinie. Mit anderen Worten, wir suchen nach dem **Achsenabschnitt** und der **Steigung**, die eine Linie definieren, die die Daten so gut wie möglich anpasst. "So gut wie möglich..." bedeutet oft, dass wir versuchen, die Summe der quadrierten **Residuen** zu minimieren, was auch als **Methode der kleinsten Quadrate (OLS)** bezeichnet wird.

* **Achsenabschnitt**: Der Wert für unsere Antwortvariable $y$, wenn unsere erklärende Variable $x=0$ ist. ($b_0$ in der obigen Formel)
* **Steigung**: Für jede Einheitserhöhung in $x$, die erwartete Zunahme oder Abnahme in $y$. Diese Definition ist nur im Fall einer einfachen linearen Regression ausreichend. Im Fall der multiplen linearen Regression müssen wir "unter Konstanthaltung aller anderen erklärenden Variablen" zu unserer Definition hinzufügen, da es mehr als eine erklärende Variable im Modell gibt. ($b_1$ in der obigen Formel)
* **Residuum**: Die Differenz zwischen einem beobachteten Wert und dem vom Modell bereitgestellten angepassten Wert: $e_i = y_i - \hat{y}_i$
* **Methode der kleinsten Quadrate (OLS)**: OLS ist eine der einfachsten und gebräuchlichsten Methoden, um die Parameter in einem linearen Regressionsmodell zu finden. Sie versucht, die Summe der quadrierten Residuen zu minimieren: $\sum_{i=1}^n e_i^2$. Eine schöne visuelle Erklärung finden Sie [hier](https://setosa.io/ev/ordinary-least-squares-regression/).

## Einfache lineare Regression mit scikit-learn

In diesem Abschnitt trainieren wir unser erstes lineares Regressionsmodell mit der Bibliothek scikit-learn. Bisher haben wir immer von Antwort- und erklärenden Variablen gesprochen. In der Data Science ist es weitaus üblicher, unsere Antwortvariable, die wir vorhersagen wollen, als **Zielvariable** zu bezeichnen. Die erklärenden Variablen werden normalerweise einfach **Features** genannt. Außerdem wird die Steigung oft als Regressions**koeffizient(en)** bezeichnet.

Scikit-learn macht es uns leicht, ihre Algorithmen zu verwenden, solange wir sicherstellen, dass unsere Daten auf die richtige Weise organisiert sind, um sie in ihre Modelle einzuspeisen.
Daher definieren wir im ersten Schritt die Zielvariable (`mpg`) und das Feature (`weight`). Nach der Instanziierung eines `LinearRegression` Objekts können wir es trainieren, indem wir die `.fit` Methode mit unseren Daten aufrufen.

Beachten Sie, dass "miles per gallon" der Kehrwert des in Deutschland weit verbreiteten "Liter pro Kilometer" ist.

```python
# Import linear regression from sklearn library
from sklearn.linear_model import LinearRegression
```

```python
# Prepare data for modeling by defining target and feature
X = cars[['weight']]  # X needs to be 2-dimensional so we need double brackets here
y = cars['mpg']
```

```python
# Show dimensions of X
X.shape
```

```python
# Instantiate our model 
lin_reg = LinearRegression()

# Train the model using our data
lin_reg.fit(X, y)
```

```python
# Intercept and slope/coefficient can be accessed via .intercept_ and .coef_
intercept = lin_reg.intercept_
slope = lin_reg.coef_[0]
print("Model intercept:", intercept)
print("Coefficient for feature weight:", slope)
```

```python
# Plotting our data points
plt.scatter(X, y)

# Adding the fitted regression line of our model
plt.plot(X, X*slope +intercept, '-', color='magenta', linewidth=2)

plt.title("Relationship between mpg and weight")
plt.ylabel("mpg");
plt.xlabel("weight");
```

In den obigen Zellen haben wir das lineare Regressionsmodell auf unseren Daten trainiert. Mit den Werten für den Achsenabschnitt und den Koeffizienten konnte die Regressionsgerade dem Streudiagramm unserer Daten hinzugefügt werden.

Wir können auch den Achsenabschnitt unseres Modells interpretieren: Laut Modell sollte `mpg` eines Autos mit einem `weight` von 0 etwa 37,28 betragen. Wie unser Fall zeigt, können Achsenabschnitte in der "realen Welt" eine Bedeutung haben oder auch nicht. Es kann kein Auto mit einem Gewicht von 0 geben...

Der Wert des Koeffizienten sagt uns, dass wir erwarten würden, dass `mpg` um etwa -5,34 abnimmt, wenn `weight` um 1 steigt.

### Quantifizierung der Modellanpassung

Bisher haben wir gesehen, wie man die Koeffizienten eines einfachen linearen Regressionsmodells schätzt. Was uns noch fehlt, ist eine geeignete Methode zur Bewertung, ob unser Modell gute Arbeit geleistet hat oder völlig nutzlos ist. Solange wir nur mit einem Feature arbeiten, können wir versuchen, die Regressionsgerade in unsere Daten zu zeichnen, um ein Gefühl für die Anpassung zu bekommen. Aber sobald wir die Anzahl der Features erhöhen, haben wir zu viele Dimensionen zum Plotten, und überhaupt ist das Plotten und Augenmaß eine ziemlich ungenaue Methode, Dinge zu messen.

#### $R^2$ (r-quadrat)

Glücklicherweise gibt es eine Möglichkeit, die Anpassung unseres Modells numerisch zu bewerten. Wir können das $R^2$ des Modells berechnen. Es hat eine einfache Interpretation: Das $R^2$ ist der *Anteil* der Varianz in der Zielvariablen (in unserem Fall mpg), der durch unsere Features erklärt werden kann (in unserem Fall haben wir nur ein Feature: weight).

Mathematisch können wir $R^2$ definieren als

$$R^2 = 1 - \frac{SSR}{SST}$$

wobei **SSR** die Summe der quadrierten Residuen ($\sum_{i=1}^n e_i^2$) und **SST** die totale Quadratsumme ($\sum_{i=1}^n (y_i - \bar{y})^2$) ist.

Der Wert für $R^2$ kann zwischen 0 und 1 liegen. Wenn das Modell die Daten perfekt erfasst, ist **SSR** gleich 0 und $R^2$ wird 1 zurückgeben. Andererseits führt eine schlechte Anpassung der Daten zu einem hohen **SSR** und das $R^2$ wird niedrig sein oder sogar gegen 0 tendieren.

#### $R^2$ in scikit-learn

Wir müssen das $R^2$ nicht von Hand berechnen. Es ist bereits in sklearn implementiert und wir können es aus dem `sklearn.metrics` Modul importieren. Wenn Sie die [sklearn Dokumentation](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html) für die `r2_score` Funktion überprüfen, können Sie sehen, dass Sie mindestens zwei Argumente übergeben müssen: *y_true* und *y_pred*.
* *y_true* sind die "wahren" Werte unserer Zielvariablen (in unserem Fall in der Variable `y` gespeichert).
* *y_pred* sind die geschätzten Zielwerte. Wir haben sie am Anfang des Notebooks $\hat{y}$ genannt. Wir könnten sie über die Steigung und den Achsenabschnitt berechnen, aber sklearn bietet uns eine bequemere Methode namens `.predict`.

```python
# Import r-squared from sklearn
from sklearn.metrics import r2_score
```

```python
# Calculate the estimated value for each data point
y_hat = lin_reg.predict(X)

# Calculate the R-squared for our model
print("R-squared:", round(r2_score(y, y_hat), 3))
```

Ein $R^2$ von 0,753 klingt zunächst gut, aber Sie sollten sich bei der Bewertung der Richtigkeit Ihres Modells nicht nur auf diese Kennzahl verlassen. Ein $R^2$ von 0,753 bedeutet "nur", dass unser Feature `weight` 75,3 % der Varianz in unserer Zielvariablen `mpg` erklärt.

Wir werden in einer späteren Sitzung darauf zurückkommen und tiefer in das Thema der ordnungsgemäßen Bewertung der Modellleistung eintauchen.

### Vorhersagen für neue Instanzen

Einer der Gründe, warum wir Modelle in der Data Science trainieren, ist natürlich, Vorhersagen für neue Beobachtungen zu treffen.
Im obigen Abschnitt haben wir bereits die `.predict` Methode verwendet, um die geschätzten y-Werte für die Daten zu berechnen, die wir zum Trainieren unseres Modells verwendet haben. Das ermöglichte es uns, das $R^2$ des Modells zu berechnen.

Wir können die `.predict` Methode auch verwenden, um einen Wert für unsere Zielvariable für eine neue Beobachtung zu schätzen (oder vorherzusagen). In unserem Fall würde dies bedeuten, dass wir den geschätzten `mpg` Wert für ein Auto mit einem bisher unbekannten Gewicht berechnen können. Um Fehler oder Warnungen zu vermeiden, müssen wir sicherstellen, dass das Format unseres neuen Datenpunkts und die Daten, die wir zum Trainieren des Modells verwendet haben, gleich sind. Daher definieren wir einen sehr kleinen Dataframe mit einer Spalte namens `weight` und einem Eintrag.

```python
# Predict the mpg value for a new car with a weight of 4.5 
new_car = pd.DataFrame({'weight': [4.5]})
y_hat_new = lin_reg.predict(new_car)
```

```python
print("Predicted value for new observation:", y_hat_new[0].round(3))
```

Laut unserem Modell hätte ein Auto mit einem Gewicht von 4,5 einen Kraftstoffverbrauch von 13,235 mpg.

## Zusammenfassung

- Lineare Regression bedeutet, die beste Anpassungslinie für Daten zu finden.
- Einfache lineare Regression umfasst eine Zielvariable und nur **ein** Feature. Multiple lineare Regression umfasst mehr als ein Feature.
- Lineare Regression kann eine Beziehung approximieren und keine Kausalität beweisen.
- Lineare Regression macht einige **Annahmen**: Linearität, Fehler mit Mittelwert Null, Strikte Exogenität, Homoskedastizität und Keine Multikollinearität.
- Die Methode der kleinsten Quadrate (OLS) ist ein Anpassungsalgorithmus, der mathematisch die beste Lösung findet.
- $\textbf{R}^2$ ist der Anteil der Variation, der durch das Modell erklärt wird.
